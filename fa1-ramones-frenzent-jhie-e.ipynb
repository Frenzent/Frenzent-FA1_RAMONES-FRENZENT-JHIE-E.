# 2. How many miles are there in 10 kilometers? Hint: there are 1.61 kilometers in a mile.

```python
# x = km
x = 10

# y = mile per km
y = 0.621371

miles = x * y

print('%d kilometers is equal to %0.2f miles.' %(x,miles))
```

```python
# TO GET THE INPUT OF THE USER

#for the input of the user 
x = float(input('Enter the distance value of the kilometer(s):'))

# conversion factor of miles and kilometers
y = 0.621371

# to calculate the miles from the input of the user:
mile = x * y

# to print the output of the code:
print('%d kilometers is equal to %0.4f miles.' %(x, mile))
```

# 3. If you run a 10 kilometer race in 42 minutes 42 seconds, what is your average pace (time per mile in minutes and seconds)? What is your average speed in miles per hour? 

```python
print("Exercises 1-2. Item 3\n")

# declare the minutes in the question 
mins = 42.7
hours = mins / 60

#x= kilometer per mile
# d = distance (km)
x = 1.61
d = 10 

# formula of km to miles
miles = d / x 

#formulas to get the pace and the miles per hour
pace = mins / miles
mph = miles / hours

print('Pace in minutes per mile:', pace)
print('Average speed in mph:', mph)
```
